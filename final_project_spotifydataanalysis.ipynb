```python
import matplotlib.pyplot as plt
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import requests


def fetch_audio_features(track_ids):
    client_credentials_manager = SpotifyClientCredentials(client_id='8043014a072944379306050f43627f9c', client_secret='ad6148c154404488bb3226dcdcfa4462')
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
    audio_features_list = []
    for track_id in track_ids:
        try:
            track = sp.audio_features(track_id)
            audio_features_list.append(track[0] if track else None)
        except requests.exceptions.ReadTimeout:
            print("Request to Spotify API timed out.")
            audio_features_list.append(None)
    return audio_features_list


def plot_radar_chart(audio_features_list, track_names):
    if len(audio_features_list) != len(track_names):
        print("Number of audio features does not match the number of track names.")
        return

    num_songs = len(audio_features_list)
    labels = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    angles += angles[:1]

    fig, axs = plt.subplots(nrows=num_songs, ncols=1, figsize=(10, 6*num_songs), subplot_kw=dict(polar=True))
    if num_songs == 1:
        axs = [axs] 

    for i, (audio_features, track_name) in enumerate(zip(audio_features_list, track_names)):
        if audio_features:
            stats = [audio_features[label] for label in labels]
            stats += stats[:1]
            axs[i].fill(angles, stats, color='blue', alpha=0.25)
            axs[i].plot(angles, stats, color='blue', linewidth=2)
            axs[i].set_yticklabels([])
            axs[i].set_xticks(angles[:-1])
            axs[i].set_xticklabels(labels)
            axs[i].set_title(track_name)
        else:
            axs[i].set_title(f"No audio features available for {track_name}")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    
    track_ids = ['27NV2KxoQ8WuLMqlTDI61F', '5l9g7py8RCblcvbZgGQgSd']  
    track_names = ['Song 1', 'Song 2']  
    audio_features_list = fetch_audio_features(track_ids)
    plot_radar_chart(audio_features_list, track_names)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials


CLIENT_ID = '8043014a072944379306050f43627f9c'
CLIENT_SECRET = '8043014a072944379306050f43627f9c'


def authenticate_spotify():
    client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
    return spotipy.Spotify(client_credentials_manager=client_credentials_manager)


def search_tracks(sp, query):
    results = sp.search(q=query, limit=5, type='track')
    tracks = results['tracks']['items']
    return tracks


def fetch_audio_features(sp, track_id):
    track = sp.audio_features(track_id)
    return track[0] if track else None


def plot_radar_chart(audio_features, track_name):
    if audio_features:
        labels = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
        stats = [audio_features[label] for label in labels]
        angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
        stats += stats[:1]
        angles += angles[:1]

        fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
        ax.fill(angles, stats, color='blue', alpha=0.25)
        ax.plot(angles, stats, color='blue', linewidth=2)
        ax.set_yticklabels([])
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(labels)
        ax.set_title(f'Radar Chart for "{track_name}"')
        plt.show()
    else:
        print("No audio features available for this track.")


def main():
    sp = authenticate_spotify()

    while True:
        query = input("Enter a song name or keyword to search (or type 'exit' to quit): ")
        if query.lower() == 'exit':
            break

        tracks = search_tracks(sp, query)

        if not tracks:
            print("No tracks found for this query.")
            continue

        print("\nSearch Results:")
        for i, track in enumerate(tracks):
            print(f"{i+1}. {track['name']} by {', '.join(artist['name'] for artist in track['artists'])}")

        selection = input("Enter the number of the track you want to see the radar chart for (or type 'back' to search again): ")
        if selection.lower() == 'back':
            continue

        try:
            selection_idx = int(selection) - 1
            if 0 <= selection_idx < len(tracks):
                track_id = tracks[selection_idx]['id']
                audio_features = fetch_audio_features(sp, track_id)
                plot_radar_chart(audio_features, tracks[selection_idx]['name'])
            else:
                print("Invalid selection. Please enter a number within the range of search results.")
        except ValueError:
            print("Invalid input. Please enter a number.")

if __name__ == "__main__":
    main()
```

```python
import plotly.graph_objs as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials


client_credentials_manager = SpotifyClientCredentials(client_id='8043014a072944379306050f43627f9c', client_secret='8043014a072944379306050f43627f9c')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)


def fetch_spotify_data(query):
    results = sp.search(q=query, type='track', limit=50)
    return results


query = "the weeknd"  
spotify_data = fetch_spotify_data(query)


track_names = [track['name'] for track in spotify_data['tracks']['items']]
popularity = [track['popularity'] for track in spotify_data['tracks']['items']]
release_dates = [track['album']['release_date'] for track in spotify_data['tracks']['items']]

# Initialize Dash app
app = dash.Dash(__name__)

# Define layout of the app
app.layout = html.Div([
    html.Div([
        dcc.Graph(id='popularity-graph'),
        html.Label('Select a track:'),
        dcc.Dropdown(
            id='track-dropdown',
            options=[{'label': track, 'value': track} for track in track_names],
            value=track_names[0]
        )
    ], style={'width': '48%', 'display': 'inline-block'}),
    
    html.Div([
        dcc.Graph(id='release-date-graph'),
        html.Label('Select a range of release dates:'),
        dcc.RangeSlider(
            id='date-range-slider',
            marks={i: release_date for i, release_date in enumerate(release_dates)},
            min=0,
            max=len(release_dates) - 1,
            value=[0, len(release_dates) - 1]
        )
    ], style={'width': '48%', 'display': 'inline-block'}),
])


@app.callback(
    Output('popularity-graph', 'figure'),
    [Input('track-dropdown', 'value')]
)
def update_popularity_graph(selected_track):
    track_index = track_names.index(selected_track)
    trace = go.Bar(x=[selected_track], y=[popularity[track_index]], name='Popularity')
    layout = go.Layout(title='Popularity of {}'.format(selected_track),
                       xaxis=dict(title='Track'),
                       yaxis=dict(title='Popularity'))
    return {'data': [trace], 'layout': layout}


@app.callback(
    Output('release-date-graph', 'figure'),
    [Input('date-range-slider', 'value')]
)
def update_release_date_graph(date_range):
    start_index, end_index = date_range
    selected_release_dates = release_dates[start_index:end_index + 1]
    trace = go.Histogram(x=selected_release_dates, name='Release Dates', marker=dict(color='orange'))
    layout = go.Layout(title='Distribution of Release Dates',
                       xaxis=dict(title='Release Date'),
                       yaxis=dict(title='Count'))
    return {'data': [trace], 'layout': layout}

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

# Nearest Neighbors (NN)

Nearest Neighbors (NN) is used in this scenario to build a recommendation system based on song features. By calculating the similarity between songs using the cosine distance metric, NN identifies the nearest neighbors to a given song, recommending similar songs to users.

Benefits:
. Personalized Recommendations: NN provides personalized song recommendations tailored to the user's preferences.
. Enhanced Music Discovery: Users can discover new songs that are similar to those they already enjoy.
. Content Curation: Music platforms can curate content more effectively by recommending songs based on their features and user interactions.

Interpretation:
The NN model analyzes the audio features of songs to identify similarities between them. It enables users to explore a diverse range of music based on their preferences, enhancing their music listening experience and driving engagement on music platforms.

```python
#nn

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors


df = pd.read_csv('D://My Documents//Desktop//tracks.csv')




audio_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[audio_features])


nn_model = NearestNeighbors(n_neighbors=10, algorithm='auto', metric='cosine')
nn_model.fit(df_scaled)


def recommend_songs(song_id, nn_model=nn_model, df=df, df_scaled=df_scaled, top_n=5):
    
    if song_id not in df['id'].values:
        print("Error: Song ID not found in the dataset.")
        return None
    
    
    idx = df.index[df['id'] == song_id][0]
    
    song_features = df_scaled[idx].reshape(1, -1)

    distances, indices = nn_model.kneighbors(song_features, n_neighbors=top_n+1)
    
    indices = indices.squeeze()[1:]
    
    similar_songs = df.iloc[indices][['id', 'name']]
    return similar_songs


example_song_id = '7xQAfvXzm3AkraOtGPWIZg'
similar_songs = recommend_songs(example_song_id)
if similar_songs is not None:
    print("Recommended songs based on similarity to the example song:")
    print(similar_songs)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
from sklearn.preprocessing import StandardScaler


df = pd.read_csv('D://My Documents//Desktop//tracks.csv')


if df.empty:
    print("Error: DataFrame is empty. Please check your dataset.")
    exit()


reader = Reader(rating_scale=(0, 100))
data = Dataset.load_from_df(df[['id_artists', 'id', 'popularity']], reader)
trainset, testset = train_test_split(data, test_size=0.2)

cf_model = SVD()
cf_model.fit(trainset)

if len(df) == 0:
    print("Error: No samples found in the dataset. Please check your data.")
    exit()


scaler = StandardScaler()
song_features = df.drop(['id', 'name', 'popularity', 'duration_ms', 'explicit', 'artists', 'id_artists', 'release_date'], axis=1)


if len(song_features.columns) == 0:
    print("Error: No features found in the dataset. Please check your data.")
    exit()

song_features_scaled = scaler.fit_transform(song_features)


if len(song_features_scaled) == 0:
    print("Error: No samples found in the dataset. Please check your data.")
    exit()

cbf_model = NearestNeighbors(n_neighbors=10, algorithm='auto', metric='cosine')
cbf_model.fit(song_features_scaled)

# Hybrid Recommendation
def recommend_songs(user_id, cf_model=cf_model, df=df, top_n=5):
    cf_predictions = cf_model.test(testset)
    cf_recommendations = [(pred.iid, pred.est) for pred in cf_predictions if pred.uid == user_id]
    cf_recommendations.sort(key=lambda x: x[1], reverse=True)
    cf_recommendations = [song_id for song_id, _ in cf_recommendations[:top_n]]
    
    return cf_recommendations

# Example usage
example_user_id = '1234'
recommendations = recommend_songs(example_user_id)
print("Collaborative filtering recommendations for user", example_user_id, ":", recommendations)
```

. Allows users to input custom audio features to refine recommendations.
. Recommends similar songs using Spotify's recommendation API based on both track ID and user-defined audio features.

Benefits:
. Customization: Users can tailor recommendations based on their preferences by inputting desired audio features.
. Enhanced Discovery: Provides personalized song recommendations, improving music discovery for users.
. Flexibility: Offers flexibility in exploring diverse music based on specific audio characteristics.

Interpretation:
This model enables users to explore and discover music that aligns with their preferences more effectively. By incorporating user-defined audio features, it offers a tailored approach to recommending songs, enhancing the overall music listening experience.
